```python
movies = pd.read_csv('/content/top10K-TMDB-movies.csv')
movies.head(20)
```

```python
movies = movies.drop(['original_language', 'popularity', 'release_date','vote_average','vote_count'], axis=1)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genre']
movies.head()
```

```python
movies = movies.drop(['overview','genre'],axis = 1)
```

```python
movies.head()
# created a new feature tags that contains information for both overview and genre as in content based recommender system goal is to compare based on their content
# tags column used for text embeddings using TF-IDF vectorization

# id column is needed as we will need to pass the poster from the api
```

```python
movies.shape
```

```python
# Converting textual data into numerical values
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000, stop_words='english')
```

```python
vector = cv.fit_transform(movies['tags'].values.astype('U')).toarray()
# converts the textual into a sparse matrix of token counts (Bag of words column) each row is a movie and each column is a word or token from volcabulary
```

```python
vector.shape
```

```python
# Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
movies[movies['title'] == 'The Godfather'].index[0] # gives the index of the movie using title
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse=True, key=lambda x:x[1])
for i in distance[0:10]:
  print(movies.iloc[i[0]].title)
# this will recommend us the top 10 movies related to The GodFather with the most similarity with their indexes
```

```python
def recommend(movie):
  index = movies[movies['title'] == movie].index[0]
  distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda x:x[1])
  for i in distance[0:10]:
    print(movies.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
